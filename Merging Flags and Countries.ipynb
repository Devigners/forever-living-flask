```python
import csv
import json

flags = {}

with open('static\\data\\flags\\us.csv', newline='') as f:
    reader = csv.reader(f)
    i = 0
    for url in reader:
        if(i != 0):
            row = url[0]
            state_name = row[row.find('Flag_of_')+len('Flag_of_'):].split('.')[0].split('_%')[0].replace('_', ' ')
            print(state_name)
            flags[state_name] = row
        i+=1

flags['Country'] = 'https://upload.wikimedia.org/wikipedia/en/a/a4/Flag_of_the_United_States.svg'
```

```python
with open("static\\data\\flags\\us.json", "w") as outfile:
    json.dump(flags, outfile)
```
