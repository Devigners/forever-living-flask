```python
gb_city_df = pd.read_csv('countries_data\\World States Cities\\GB\\GB.csv')
gb_city_df
```

```python
cities_county = gb_city_df[['city_ascii', 'admin_name']]
```

```python
cities_county.to_csv("countries_data\\World States Cities\\GB\\GB data.csv", index=False)
```

```python
with open("static\\data\\countries\\gb data.txt", "r") as file:
    result = [[x.strip() for x in line.split(', ')] for line in file]

states = []
for r in result:
    if r[0] not in states:
        states.append(r[0])
        
states = list(set(states))
for state in states:
    print(state)
```
