```python
import pandas as pd

states = []
```

```python
us_city_df = pd.read_csv('countries_data\\World States Cities\\US\\US.csv')
us_city_df
```

```python
states = states + us_city_df['admin_name'].to_list()
```

```python
cities_states = us_city_df[['admin_name', 'city_ascii']]
cities_states 
```

```python
us_county_df = pd.read_csv('countries_data\\World States Cities\\USA counties\\uscounties.csv')
us_county_df.rename(columns = {'state_name':'state', 'county_ascii':'county'}, inplace = True)
us_county_df
```

```python
states = states + us_county_df['state'].to_list()
```

```python
us_neighborhood_df = pd.read_csv('countries_data\\World States Cities\\USA neighborhoods\\usneighborhoods.csv')
us_neighborhood_df.rename(columns = {'state_name':'state', 'city_name':'city', 'neighborhood_ascii':'neighborhood'}, inplace = True)
us_neighborhood_df
```

```python
states = states + us_neighborhood_df['state'].to_list()
```

```python
states = list(set(states))
```

```python
us_df = pd.DataFrame(columns=['state', 'city', 'county', 'neighborhood'])
```

```python
us_df['state'] = states
```

```python
us_df = pd.concat([us_df, cities_states, us_county_df, us_neighborhood_df])
```

```python
us_df.drop_duplicates(inplace=True)
us_df
```

```python
us_df.to_csv("countries_data\\World States Cities\\US\\US data.csv", index=False)
```

```python
def findLocalities(required_country = 'United States'):
    countries_codes = {'Australia': 'AU', 'United States':'US', 'Canada':'CA', 'Great Britain':'GB'}
    required_country_code = countries_codes[required_country]
    data = pd.read_csv('static\\data\\countries\\'+required_country_code+' data.csv').values.tolist()

    localities = []
    for row in data:
        entry = []
        for item in row:
            if isinstance(item, str):
                entry.append(item)
        localities.append(entry)

    country_localities = []
    for locality in localities:
        # add your search url here
        country_localities.append({'name': ', '.join(locality), 'url': '#'})
    
    return country_localities
```
